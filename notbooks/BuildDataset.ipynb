```python
import pandas as pd
import numpy as np

from collections import Counter
```

## Wczytanie bazowego datasetu

```python
df = pd.read_csv("glove_df_dbpedied_p.csv")
```

```python
df
```

## W poprzednich skryptach doszło do utworzenia duplikatów w kolumnie text, więc teraz się ich pozbywamy

```python
df = df.drop_duplicates(subset=["text"]); df
```

## Zbudowanie listy klas w obrębie których będziemy dokonywać predykcji

```python
classes = list(df.columns.values[3:-1])
classes
```

## Sprawdzenie rozkładu klas
Chodzi o podliczenie ile mamy próbek odpowiadających danej klasie. Jeżeli mamy tylko jedną próbkę danej klasy to dodajemy ID takiego wiersza do tablicy

```python
rozklad = Counter()

for cc, row in df.iterrows():
    cat = classes[np.array(row[classes]).argmax()]
    rozklad[cat] += 1

theonly = []
rev = reversed(rozklad.most_common())
for item in rev:
    if item[1] == 1:
        theonly.append(item[0])
```

```python
rozklad.most_common()
```

## Wypisanie klas, ktore mają tylko jeden przykład

```python
theonly
```

## Usuwamy kolumny, które mają tylko pojedyńcze wystąpienia klas

```python
for item in theonly:
    classes.remove(item)
    
df = df.drop(columns=theonly)
```

```python
classes
```

## Do zbioru testowego dodajemy próbki, które odpowiadają każdej z klasy
Dokładnie to po int(0.2*ilość próbek)+1

```python
rowtest = []
rowtestcats = Counter()

for cc, row in df.iterrows():
    cat = classes[np.array(row[classes]).argmax()]
    
    if row[cat] == 1:
        if rowtestcats[cat] < int(0.2*rozklad[cat])+1:
#             print(cc, cat)
            rowtestcats[cat] += 1
            rowtest.append(cc)
```

```python
len(rowtestcats), len(classes), len(rowtest)
```

## Utworzenie nowego dataframe

```python
testds = pd.DataFrame()
for row in rowtest:
    testds = testds.append(df.loc[row])
```

## Usunięcie niepotrzebnej kolumny

```python
testds = testds.drop(columns=["Unnamed: 0"])
```

```python
testds[classes] = testds[classes].astype(int)
```

```python
testds
```

## Utworzenie zbioru uczącego

```python
trainds = pd.DataFrame()

for row in range(len(df)):
    if row not in rowtest:
        trainds = trainds.append(df.iloc[row])
```

## Usunięcie niepotrzebnej kolumny

```python
trainds = trainds.drop(columns=["Unnamed: 0"])
```

```python
trainds[classes] = trainds[classes].astype(int)
```

```python
trainds.head()
```

## Zapis do pliku

```python
testds.to_csv("test_p.csv")
trainds.to_csv("train_p.csv")
```
